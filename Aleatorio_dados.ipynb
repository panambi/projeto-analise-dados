```python
# importação de bibliotecas
import random
import matplotlib.pyplot as plt

# Gerar uma lista com 10 números aleatorios entre 1 e 100
dados = [random.randint(1, 100) for _ in range(10)]

#Exibir a lista gerada
print("Lista de números aleatórios:", dados)

# Plotar os dados
plt.plot(dados, marker='o')   # marcador em cada dados para melhor visualização
plt.xlabel('Índice')
plt.ylabel('Valor')
plt.title('Gráfico de Dados Aleatórios')
plt.grid(True)
plt.show()
```
